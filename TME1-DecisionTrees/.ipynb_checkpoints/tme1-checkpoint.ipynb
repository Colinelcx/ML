# Exercice 1 - Entropie

### Question 1.1

```python
def entropie(vect):
    """ 
        Calcule l'entropie du vecteur vect
    """
    size = len(vect)
    cnt = Counter(vect)
    h = 0 # entropie
    for y in cnt.keys():
        p = cnt[y] / size
        h -= p * np.log(p)
    return h
```

```python
v0 = [1,1,1,1,1,1]
v1 = [1,2,1,1,1,1]
v2 = [1,1,2,1,2,2]
print('entropie de v0 : h(Y)=', entropie(v0))
print('entropie de v1 : h(Y)=', entropie(v1))
print('entropie de v2 : h(Y)=', entropie(v2))
```

### Question 1.2

```python
def entropie_cond(list_vect):
    """
        Calcule l'entropie conditionelle de la partition à partir de la liste 
        des labels (liste_vect)
    """
    sum_pj = np.array(list_vect).size
    h = 0 # entropie conditionnelle
    for vect_i in list_vect :
        p = len(vect_i) / sum_pj
        h += p * entropie(vect_i)
    return h
```

```python
print('entropie conditionelle : h(Y|P)=', entropie_cond([v0,v1,v2]))
```

### Question 1.3

```python
# chargement d'un extrait de la base imdb

# data : tableau (films ,features), id2titles : dictionnaire  id -> titre ,
# fields : id  feature  -> nom
[data , id2titles , fields]= pickle.load(open("imdb_extrait.pkl","rb")) # la  derniere  colonne  est le vote
datax = data[: ,:32]
datay = np.array([1 if x[33] >6.5  else  -1 for x in data])
```

```python
entropie_attr = []
entropie_cond_attr = []

for attr in range(datax.shape[1]):
    entropie_attr.append(entropie(datay))
    entropie_cond_attr.append(entropie_cond([datay[(datax[:, attr] == 1)],datay[(datax[:, attr] != 1)]]))
    print("attribut", fields[attr])
    print("\tentropie =", entropie_attr[attr])
    print("\tentropie conditonnelle =", entropie_cond_attr[attr])
    print('\tdifference = ', entropie_cond_attr[attr] - entropie_attr[attr])
```

Pour l'entropie, une valeur de 0 signifie que les valeurs sont toutes identiques. Plus l'entropie augmente, plus le désordre est grand. Une valeur de 1 signifique que les valeurs sont équitablement distribuées. 

L'entropie conditionnelle d'une partition mesure l'homogénéité d'une partition obtenue. Une entropie de 0 correspond à une séparation parfaite (un seul label est présent dans chacune des partitions), et celle-ci est maximale lorsque l’ensemble est le plus désordonné possible.

Pour la première partition, l'attribut est donc celui ayant la plus faible différence entre l'entropie conditionnelle induite par l'attribut et l'entropie, car on souhaite avoir une séparation la moins désorbonnée possible.

```python
i = np.argmin(np.array(entropie_cond_attr) - np.array(entropie_attr))
print("Meilleur attribut pour la partition :",fields[i])
```

Le meilleur attribut pour la première partition est donc l'attribut "Drama".

### Question 1.6

Cependant, ces scores ne sont pas des indicateurs fiables, car on teste sur les données avec lesquelles on a appris. Lorsque la profondeur est trop grande, on fait du sur-apprentissage, et si on teste avec de nouvelles données, on risque d'obtenir de mauvais résultats.

Pour produire un bon modèle, il faut séparer les données en un ensemble d'apprentissage et un ensemble de test.

# Sur et sous apprentissage

### Question 1.7

```python
def split_data(datax, datay, perc):
    """ 
        Sépare les données en un ensemble d'apprentissage et un ensemble de test
        perc : pourcentage de données dans l'ensemble d'apprentissage
    """
    size = datax.shape[0] # nombre de données
    indices = np.arange(size)
    np.random.shuffle(indices) # mélange aléatoire des indices
    nb_app = int(size*perc) # nombre de données d'appentissage
    id_app = indices[0:nb_app] # indices des données d'apprentissage
    id_test = indices[nb_app:] # indice des données de test
    return datax[id_app], datax[id_test], datay[id_app], datay[id_test]
```

```python
datax_app, datax_test, datay_app, datay_test = split_data(datax, datay, 0.8)
print(datax_app.shape, datax_test.shape, datay_app.shape, datay_test.shape)
```

```python
def courbes_erreurs(datax, datay, perc):
    """ 
        Trace les courbes de l'erreur en apprentissage et de l'erreur en test 
        en fonction de la profondeur du modèle (de 2 à 30)
    """
    depths = np.arange(28) + 2
    datax_app, datax_test, datay_app, datay_test = split_data(datax, datay, perc)
    error_app = []
    error_test = []
    
    for depth in depths:
        dt = DecisionTree()
        dt.max_depth = depth # on fixe la  taille  de l’arbre a depth
        dt.min_samples_split = 2 # nombre  minimum d’exemples  pour  spliter  un noeud
        dt.fit(datax_app, datay_app) # apprentissage
        error_app.append(1 - dt.score(datax_app ,datay_app)) # erreur en apprentissage        
        error_test.append(1 - dt.score(datax_test ,datay_test)) # erreur en test
    
    plt.figure()
    plt.title("Courbes d'erreurs avec taux d'apprentissage="+str(perc))
    plt.plot(depths, error_app, label="Erreur d'apprentissage", color='cornflowerblue')
    plt.xlabel("profondeur de l'arbre")
    plt.plot(depths, error_test, label="Erreur de test", color='crimson')
    plt.ylabel("erreur")
    plt.legend()
    plt.show()
```

```python
courbes_erreurs(datax, datay, 0.5)
```

```python
courbes_erreurs(datax, datay, 0.8)
```

```python
courbes_erreurs(datax, datay, 0.98)
```

### Question 1.8

Lorsque $taux=0.02$, les résultats sont bons en apprentissage mais mauvais en test (le taux d'erreur est supérieur à 0.35). Etant donné qu'il y a peu d'exemples d'apprentissage, le modèle apprend quasiment par coeur sur ceux-ci, mais il n'a pas assez récolté d'informations pertinentes pour classifier correctement en test.

Lorsqu'on a plus de données d'apprentissage ($taux=0.5$ ou $taux=0.8$), les résultats sont meilleurs en test. Mais lorsque l'on augmente trop la profondeur de l'arbre, le taux d'erreur diminue sur l'ensemble d'apprentissage, mais pas sur l'ensemble de test. Il tend même à se dégrader à partir de $n=5$ où $n$ est la profondeur de l'arbre. C'est le problème de sur-apprentissage : l'arbre se concode trop à l'ensemble d'apprentissage et fait des erreurs sur l'ensemble de test. 

Lorsque $taux=0.95$, les résultats en test sont bons mais peu fiables car il y a peu de données de test.


### Question 1.9

On dispose de peu de données, et nous n'avons pas d'ensemble de test. On a partionné nos données afin de créer un ensemble de test, mais au passage nous avons perdu de l'information lors de la production de notre modèle. Pour palier à ce problème, on peut améliorer notre classifieur en utilisant la procédure de séléction par validation croisée. En effet, cette technique nous permet d'utiliser nos données à la fois en apprentissage et en test, et donc de ne pas perdre d'échantillons lors de l'aprentissage.

# Validation croisée : sélection de modèle

```python
def partionner(N, n):
    """ 
        Partionne les N données en n ensembles d'apprentissage
        Retourne n ensembles d'indices
        Attention : n doit être un diviseur de N
    """
    
    indices = np.arange(N)
    np.random.shuffle(indices)
    return indices.reshape((n, int(N/n)))
```

```python
datax.shape[0]
```

```python
ensembles = partionner(datax.shape[0], 11)
print(ensembles.shape)
```

```python
def validation_croisee(datax, datay, n, depth, min_sample_split=2):
    """
        Calcul le meilleur modèle en utilisant la procèdure de validation croisée
        - datax, datay : données d'apprentissage et de test
        - n : nombre d'ensembles d'apprentissage
        - depth : profondeur de l'arbre
        - min_sample_split : nombre minimum d'exemples pour spliter
        Retourne le modèle, son erreur moyenne et l'erreur en test
    """
    best_dt = DecisionTree() # meilleur modèle
    best_err_moy = 2 # erreur moyenne
    err_test = 0
    indices_ens = partionner(datax.shape[0], n) # ensembles d'apprentissage
    for ens_i in indices_ens:
        datax_app = np.delete(datax, ens_i, axis=0)
        datay_app = np.delete(datay, ens_i, axis=0)
        
        dt = DecisionTree()
        dt.max_depth = depth # on fixe la  taille  de l’arbre a depth
        dt.min_samples_split = min_sample_split # nombre  minimum d’exemples  pour  spliter  un noeud
        dt.fit(datax_app, datay_app) # apprentissage
        
        err_test = 1 - dt.score(datax[ens_i], datay[ens_i]) # erreur en test
        
        # calcul de l'erreur moyenne
        err_moy = 0        
        for ens in indices_ens:
            err_moy += 1 - dt.score(datax[ens], datay[ens])
        err_moy = err_moy / len(indices_ens)
        
        if err_moy < best_err_moy :
            best_dt = dt
            best_err_moy = err_moy
            best_err_test = err_test
            
    return best_dt, best_err_moy, best_err_test
    
```

Test de la validation croisée :

```python
dt, err_moy, err_test = validation_croisee(datax, datay, 11, 4)
print("-- Validation croisée --")
print("\terreur moyenne = ",err_moy)
print("\terreur en test = ", err_test)
dt.to_png("img/exempleVC.png",fields)
img = Image(filename='img/exempleVC.png') 
display(img)
```

### Expériences avec validation croisée

```python
def courbes_erreurs_vc(datax, datay, n):
    """ 
        Trace les courbes de l'erreur en apprentissage et de l'erreur en test 
        en fonction de la profondeur du modèle (de 2 à 30) en validation croisée
        n = nombre d'ensembles d'apprentissage
    """
    depths = np.arange(20) + 2
    error_moy = []
    error_test = []
    
    for depth in depths:
        dt, err_moy, err_test = validation_croisee(datax, datay, n, depth)
        error_moy.append(err_moy) # erreur moyenne   
        error_test.append(err_test) # erreur en test
    
    plt.figure()
    plt.title("Courbes d'erreurs avec "+str(n)+" ensembles d'apprentissage")
    plt.plot(depths, error_moy, label="Erreur moyenne", color='forestgreen')
    plt.plot(depths, error_test, label="Erreur en test", color='slateblue')
    plt.xlabel("profondeur de l'arbre")
    plt.ylabel("erreur")
    plt.legend()
    plt.show()
```

```python
courbes_erreurs_vc(datax, datay, 3)
```

```python
courbes_erreurs_vc(datax, datay, 11)
```

```python
courbes_erreurs_vc(datax, datay, 33)
```

La validation permet d'obtenir de meilleurs résultats en test (le taux d'erreur en test descend jusqu'à 0.25 voire 0.18) comparé aux expériences précédentes. Ceci est dû au fait qu'on teste sur tous les ensembles de données avant de choisir le meilleur modèle.
